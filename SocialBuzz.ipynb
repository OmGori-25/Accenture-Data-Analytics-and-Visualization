```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
content_df = pd.read_csv('/content/drive/MyDrive/Accenture Forage/Content (1).csv',index_col=0)
content_df
```

```python
content_df.info()
```

```python
content_df.describe()
```

We do not need User Id and URL for analysis of top 5 categories, so
drop the columns of User Id and URL.

```python
content_df.drop(labels = ['User ID','URL'],axis = 1 ,inplace = True)
```

```python
content_df.info()
```

```python
content_df['Category'].value_counts()
```

We will need to clean up categories having punctuation and capitalization.



```python
content_df['Category'] = content_df['Category'].apply(lambda x : x.strip('"').lower())
```

```python
content_df['Category'].value_counts()
```

```python
content_df.rename(columns={'Type':'Content_Type'},inplace = True)
content_df['Content_Type'].value_counts()
```

```python
reactions_df = pd.read_csv('/content/drive/MyDrive/Accenture Forage/Reactions (1).csv',index_col=0)
reactions_df
```

```python
reactions_df.info()
```

```python
reactions_df.describe()
```

We have to drop User ID column since its not needed in analysis, we also need to drop the missing value rows and change Datetime column to a datetime data type.

```python
reactions_df.drop(labels = ['User ID'],axis = 1 ,inplace = True)
reactions_df
```

```python
reactions_df['Datetime'] = pd.to_datetime(reactions_df['Datetime'])
reactions_df
```

```python
reactions_df.dropna(inplace = True)
```

```python
reactions_df.rename(columns={'Type':'Reaction_Type'},inplace = True)
```

```python
reactions_df.info()
```

```python
reactiontypes_df = pd.read_csv('/content/drive/MyDrive/Accenture Forage/ReactionTypes (1).csv',index_col=0)
reactiontypes_df
```

```python
reactiontypes_df.info()
```

```python
reactiontypes_df.rename(columns={'Type':'Reaction_Type'},inplace = True)
reactiontypes_df
```

## **Merge 3 tables for analysis**

```python
top_performers = reactions_df.merge(content_df,on='Content ID')
top_performers
```

```python
top_performers = top_performers.merge(reactiontypes_df,on='Reaction_Type')
top_performers
```

## **Top Category Analysis**

```python
# top_performers.drop(labels=['Datetime'],axis = 1 ,inplace = True)
top_five = top_performers.loc[:,['Category','Score']].groupby('Category').sum().sort_values(by='Score',ascending = False)
top_five
```

**Top five categories by aggregate popularity area animals , science , healthy eating , technology,food **
